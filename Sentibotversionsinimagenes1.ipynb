```python
import tensorflow as tf

# Parámetros
ancho_imagen = 48
alto_imagen = 48
tamanio_lote = 32
nombres_clases = ['enojado', 'miedo', 'feliz', 'neutral', 'triste', 'sorpresa']
num_clases = len(nombres_clases)
epocas = 50

# Rutas
dir_entrenamiento = "/content/images/train"
dir_validacion = "/content/images/validation"

# Dataset de entrenamiento
entrenamiento = tf.keras.utils.image_dataset_from_directory(
    dir_entrenamiento,
    labels="inferred",
    label_mode="categorical",
    image_size=(ancho_imagen, alto_imagen),
    batch_size=tamanio_lote,
    color_mode="grayscale",
    shuffle=True
)

# Dataset de validación
validacion = tf.keras.utils.image_dataset_from_directory(
    dir_validacion,
    labels="inferred",
    label_mode="categorical",
    image_size=(ancho_imagen, alto_imagen),
    batch_size=tamanio_lote,
    color_mode="grayscale",
    shuffle=True
)
```

```python
import os

def contar_imagenes(directorio):
    total = 0
    print(f"Conteo de imágenes en '{directorio}':")
    for clase in os.listdir(directorio):
        ruta_clase = os.path.join(directorio, clase)
        if os.path.isdir(ruta_clase):
            num_imgs = len([f for f in os.listdir(ruta_clase) if f.lower().endswith(('.jpg','.jpeg','.png'))])
            print(f"  {clase}: {num_imgs} imágenes")
            total += num_imgs
    print(f"Total: {total} imágenes\n")
    return total

# Contar imágenes
total_train = contar_imagenes(dir_entrenamiento)
total_validacion = contar_imagenes(dir_validacion)
```

### 4. Modelo CNN:

Definimos la arquitectura del modelo CNN

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Dense, Flatten, Dropout, BatchNormalization, Activation
from tensorflow.keras.optimizers import Adam

# Definir modelo
modelo = Sequential()

## Extracción de características
modelo.add(Conv2D(32, (3,3), padding='same', input_shape=(alto_imagen, ancho_imagen, 1)))
modelo.add(BatchNormalization())
modelo.add(Activation('relu'))
modelo.add(MaxPool2D(pool_size=(2,2)))
modelo.add(Dropout(0.2))

modelo.add(Conv2D(64, (5,5), padding='same'))
modelo.add(BatchNormalization())
modelo.add(Activation('relu'))
modelo.add(MaxPool2D(pool_size=(2,2)))
modelo.add(Dropout(0.2))

modelo.add(Conv2D(128, (3,3), padding='same'))
modelo.add(BatchNormalization())
modelo.add(Activation('relu'))
modelo.add(MaxPool2D(pool_size=(2,2)))
modelo.add(Dropout(0.2))

modelo.add(Conv2D(256, (3,3), padding='same'))
modelo.add(BatchNormalization())
modelo.add(Activation('relu'))
modelo.add(MaxPool2D(pool_size=(2,2)))
modelo.add(Dropout(0.2))

modelo.add(Conv2D(512, (3,3), padding='same'))
modelo.add(BatchNormalization())
modelo.add(Activation('relu'))
modelo.add(MaxPool2D(pool_size=(2,2)))
modelo.add(Dropout(0.2))

## Clasificación
modelo.add(Flatten())
modelo.add(Dense(128))
modelo.add(BatchNormalization())
modelo.add(Activation('relu'))
modelo.add(Dropout(0.2))

modelo.add(Dense(256))
modelo.add(BatchNormalization())
modelo.add(Activation('relu'))
modelo.add(Dropout(0.2))

modelo.add(Dense(num_clases, activation='softmax'))

# Revisar resumen del modelo
modelo.summary()

# Compilamos
optimizador = Adam(learning_rate=1e-4, decay=1e-4 / epocas)
modelo.compile(loss='categorical_crossentropy', optimizer=optimizador, metrics=['accuracy'])
```

```python
import tensorflow as tf
from tensorflow.keras import layers
import os

# -----------------------------
# Parámetros
# -----------------------------
ancho_imagen = 48
alto_imagen = 48
tamanio_lote = 32
nombres_clases = ['enojado', 'miedo', 'feliz', 'neutral', 'triste', 'sorpresa']
num_clases = len(nombres_clases)
epocas = 50

# Rutas
dir_entrenamiento = "/content/images/train"       # tu carpeta de entrenamiento
dir_validacion = "/content/images/validation"    # tu carpeta de validación

# -----------------------------
# Data augmentation
# -----------------------------
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
    layers.RandomContrast(0.1),
])

# -----------------------------
# Dataset de entrenamiento
# -----------------------------
entrenamiento = tf.keras.utils.image_dataset_from_directory(
    dir_entrenamiento,
    labels="inferred",
    label_mode="categorical",
    image_size=(ancho_imagen, alto_imagen),
    batch_size=tamanio_lote,
    color_mode="grayscale",
    shuffle=True
)

# -----------------------------
# Dataset de validación
# -----------------------------
validacion = tf.keras.utils.image_dataset_from_directory(
    dir_validacion,
    labels="inferred",
    label_mode="categorical",
    image_size=(ancho_imagen, alto_imagen),
    batch_size=tamanio_lote,
    color_mode="grayscale",
    shuffle=True
)

# -----------------------------
# Normalización de píxeles
# -----------------------------
normalizacion = layers.Rescaling(1./255)

entrenamiento = entrenamiento.map(lambda x, y: (normalizacion(x), y))
validacion = validacion.map(lambda x, y: (normalizacion(x), y))

# -----------------------------
# Aplicar data augmentation solo al entrenamiento
# -----------------------------
AUTOTUNE = tf.data.AUTOTUNE
entrenamiento = entrenamiento.map(lambda x, y: (data_augmentation(x, training=True), y),
                                   num_parallel_calls=AUTOTUNE).prefetch(AUTOTUNE)
validacion = validacion.prefetch(AUTOTUNE)

# -----------------------------
# Verificar número de imágenes
# -----------------------------
def contar_imagenes(directorio):
    total = 0
    print(f"Conteo de imágenes en '{directorio}':")
    for clase in os.listdir(directorio):
        ruta_clase = os.path.join(directorio, clase)
        if os.path.isdir(ruta_clase):
            num_imgs = len([f for f in os.listdir(ruta_clase) if f.lower().endswith(('.jpg','.jpeg','.png'))])
            print(f"  {clase}: {num_imgs} imágenes")
            total += num_imgs
    print(f"Total: {total} imágenes\n")
    return total

total_train = contar_imagenes(dir_entrenamiento)
total_validacion = contar_imagenes(dir_validacion)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Dense, Flatten, Dropout, BatchNormalization, Activation
from tensorflow.keras.optimizers import Adam

# Definir modelo
modelo = Sequential()

## Extracción de características
modelo.add(Conv2D(32, (3,3), padding='same', input_shape=(alto_imagen, ancho_imagen, 1)))
modelo.add(BatchNormalization())
modelo.add(Activation('relu'))
modelo.add(MaxPool2D(pool_size=(2,2)))
modelo.add(Dropout(0.2))

modelo.add(Conv2D(64, (5,5), padding='same'))
modelo.add(BatchNormalization())
modelo.add(Activation('relu'))
modelo.add(MaxPool2D(pool_size=(2,2)))
modelo.add(Dropout(0.2))

modelo.add(Conv2D(128, (3,3), padding='same'))
modelo.add(BatchNormalization())
modelo.add(Activation('relu'))
modelo.add(MaxPool2D(pool_size=(2,2)))
modelo.add(Dropout(0.2))

modelo.add(Conv2D(256, (3,3), padding='same'))
modelo.add(BatchNormalization())
modelo.add(Activation('relu'))
modelo.add(MaxPool2D(pool_size=(2,2)))
modelo.add(Dropout(0.2))

modelo.add(Conv2D(512, (3,3), padding='same'))
modelo.add(BatchNormalization())
modelo.add(Activation('relu'))
modelo.add(MaxPool2D(pool_size=(2,2)))
modelo.add(Dropout(0.2))

## Clasificación
modelo.add(Flatten())
modelo.add(Dense(128))
modelo.add(BatchNormalization())
modelo.add(Activation('relu'))
modelo.add(Dropout(0.2))

modelo.add(Dense(256))
modelo.add(BatchNormalization())
modelo.add(Activation('relu'))
modelo.add(Dropout(0.2))

modelo.add(Dense(num_clases, activation='softmax'))

# Revisar resumen del modelo
modelo.summary()

# Compilamos
optimizador = Adam(learning_rate=1e-4, decay=1e-4 / epocas)
modelo.compile(loss='categorical_crossentropy', optimizer=optimizador, metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# Definir EarlyStopping
early_stopping = EarlyStopping(
    monitor='val_loss',      # monitoriza la pérdida de validación
    patience=8,              # espera 5 épocas sin mejora antes de detenerse
    restore_best_weights=True # restaura los pesos del mejor epoch
)

# Entrenamiento con EarlyStopping
history = modelo.fit(
    entrenamiento,
    validation_data=validacion,
    epochs=150,
    callbacks=[early_stopping]  # <-- aplicamos early stopping
)
```
